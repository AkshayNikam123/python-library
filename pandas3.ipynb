```python
d1={'key1':[1,2,3,4,5,6],
    'key2':[7,8,9,10,11,12],
    'key3':[13,14,15,16,17,18],
    'key4':[19,20,21,22,23,24]
   }
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(d1) # d1 should have same length in valu
```

```python
pd.DataFrame(d1)
```

```python
d2={'key1':[1,2,3,4,5,6],
    'key2':[7,8,9,10,'akshay',12],
    'key3':[13,14,15,16,'nikam',18],
    'key4':[19,20,21,22,23,24]
   }
```

```python
df2=pd.DataFrame(d2)
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='inner',on='key1') #merge on common column key1 elements #inner for common elements from both
```

```python
pd.merge(df1,df2,how='left',on='key4')
```

```python
df1.join(df2)
```
